```python
df['discounted_price'] = df['discounted_price'].str.replace('‚Çπ','').str.replace(',','').astype(float)
```

```python
df['actual_price'] = df['actual_price'].str.replace('‚Çπ','').str.replace(',','').astype(float)
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype(float)/100
```

```python
df.dtypes
```

```python
df.sort_values(by = 'rating',ascending = True)
```

There is a special characer '|' in the rating column of row 1279. Lets's check if there are more than one row that has special character

```python
df[df['rating'].str.contains(r'\|')]
```

```python
df = df[~df['rating'].str.contains(r'\|')]
```

df[~df['rating'].str.contains(r'\|')]

uses the negated boolean mask to select the rows from the DataFrame where the 'rating' column does not contain the '|' character. This filters out the rows with the '|' character from the DataFrame.

```python
df['rating'] = df['rating'].astype(float)
```

```python
df['rating_count'] = df['rating_count'].str.replace(',','').astype(float)
```

```python
df.dtypes
```

Lets create a weighted_rating for the following resons:

1. Account for Rating Count: The 'rating_count' column represents the number of ratings a product has received. Products with a higher number of ratings often provide more reliable information about their quality. By incorporating rating counts into the weighted rating calculation, you can give more weight to products with a significant number of ratings, which helps mitigate the influence of outliers or products with very few ratings.
2. Improve Accuracy: Averaging ratings alone may not accurately represent the overall quality of a product, especially when there are products with high average ratings but very few ratings. By considering the number of ratings in conjunction with the average rating, a weighted rating can provide a more reliable measure of a product's quality.
3. Better Decision Making: When comparing products based on their ratings, considering the weighted rating can help you make more informed decisions. It allows you to prioritize products with both high average ratings and a sufficient number of ratings, indicating a combination of quality and popularity.
4. Enhanced Customer Experience: Providing a weighted rating can be beneficial for customers browsing through the products. It helps them quickly assess the overall quality and popularity of a product, making it easier for them to make purchasing decisions.

```python
df['weighted_rating'] = df['rating'] * df['rating_count']
```

Lets classify the category column in to 'major category' and 'sub categories' to identify the products more easily

```python
df['major_category'] = df['category'].str.split('|').str[0]
```

```python
df['sub-catagories'] = df['category'].str.split('|').str[-1]
```

# EDA and Visualization

Exploratory Data Analysis (EDA) is a critical step in the data analysis process. Here are several reasons why EDA is essential:

1. Understand the Data: EDA helps in gaining a deep understanding of the dataset at hand. It allows us to examine the structure, content, and quality of the data. By exploring the data, we can identify any missing values, outliers, or inconsistencies that may need to be addressed.

2. Identify Patterns and Relationships: EDA enables us to uncover patterns, trends, and relationships within the data. By visualizing and summarizing the data, we can identify variables that are strongly correlated, discover hidden patterns, or detect outliers that may impact subsequent analysis or modeling.

Here we are going to find out how our product are distributed across the categories, product reviews and ratings.

```python
import matplotlib.pyplot as plt
```

```python
major_category_counts = df['major_category'].value_counts()[:30] # Select only the top 30 main categories.
```

# Word Cloud

```python
from wordcloud import WordCloud
```

```python
# Analyze the reviews by creating word clouds or frequency tables of the most common words used in the reviews.
reviews_text = ' '.join(df['review_content'].dropna().values)
wordcloud = WordCloud(width=800, height=800, background_color='white', min_font_size=10).generate(reviews_text)
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

Word clouds visually representing the most prominent words in a text, it can help identify the main themes, topics, or keywords within the data. The larger and more prominent words in the cloud represent the most frequently mentioned or important concepts.

## **Let's see the correlation between the features. But first we need to drop the non-numneric columns**




```python
# Drop non-numeric columns
numeric_cols = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_cols.corr()

# Print the correlation matrix
print(correlation_matrix)
```

```python
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

1. Discounted Price and Actual Price: The variables "discounted_price" and "actual_price" have a strong positive correlation of 0.961910. This indicates that as the discounted price increases, the actual price tends to increase as well. It suggests that discounts are applied to products with higher original prices.
2. Discount Percentage: The variable "discount_percentage" has a negative correlation with "discounted_price" (-0.242298) and "actual_price" (-0.117855). This implies that as the discount percentage increases, the discounted price and actual price tend to decrease. It confirms the intuitive relationship that higher discounts result in lower prices.
